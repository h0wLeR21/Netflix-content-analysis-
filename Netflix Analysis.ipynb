```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn  as sns
```

```python
df = pd.read_csv('netflix_titles.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
cleaned_df = df.copy()
```

```python
cleaned_df['country'] = cleaned_df['country'].fillna('Unknown')
cleaned_df['director'] = cleaned_df['director'].fillna('Unknown')
cleaned_df['cast'] = cleaned_df['cast'].fillna('Unknown')
most_common_date = cleaned_df['date_added'].mode()[0]
cleaned_df['date_added'] = cleaned_df['date_added'].fillna(most_common_date)
most_common_rating = cleaned_df['rating'].mode()[0]
cleaned_df['rating'] = cleaned_df['rating'].fillna(most_common_rating)
cleaned_df['duration'] = cleaned_df['duration'].fillna('Unknown')
```

```python
plt.figure(figsize=(6,4))
df['type'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Movies vs TV Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(6,4))
df['type'].value_counts().plot(
    kind='bar',
    color='#E50914',  # Netflix red
    edgecolor='white' # white edges to pop
)

plt.title('Movies vs TV Shows on Netflix', color='white', fontsize=14)
plt.xlabel('Type', color='white')
plt.ylabel('Count', color='white')

# Dark background
plt.gca().set_facecolor('#000000')  # plot area
plt.gcf().patch.set_facecolor('#000000')  # figure background

# White ticks & spines
plt.xticks(color='white')
plt.yticks(color='white')
for spine in plt.gca().spines.values():
    spine.set_color('white')

plt.grid(axis='y', color='gray', alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(10,5))
df['release_year'].value_counts().sort_index().plot(kind='line', marker='o')
plt.title('Number of Titles Released Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Count of Titles')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
df['release_year'].value_counts().sort_index().plot(
    kind='line',
    marker='o',
    color='#E50914',  # Netflix red
    linewidth=2,
    markersize=6
)

plt.title('Number of Titles Released Over the Years', color='white', fontsize=14, fontweight='bold')
plt.xlabel('Release Year', color='white')
plt.ylabel('Count of Titles', color='white')

# Dark background
plt.gca().set_facecolor('#000000')
plt.gcf().patch.set_facecolor('#000000')

# White ticks & spines
plt.xticks(color='white')
plt.yticks(color='white')
for spine in plt.gca().spines.values():
    spine.set_color('white')

plt.grid(True, color='gray', alpha=0.3)
plt.show()
```

```python
genres = df['listed_in'].str.split(', ').explode()
top_genres = genres.value_counts().head(10)
plt.figure(figsize=(8,5))
top_genres.plot(kind='barh', color='orchid', edgecolor='black')
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.gca().invert_yaxis()
plt.show()
```

```python

genres = df['listed_in'].str.split(', ').explode()
top_genres = genres.value_counts().head(10)

plt.figure(figsize=(8,5))
top_genres.plot(
    kind='barh',
    color='#E50914',   # Netflix red
    edgecolor='white'  # white edges for contrast
)

plt.title('Top 10 Genres on Netflix', color='white', fontsize=14, fontweight='bold')
plt.xlabel('Count', color='white')
plt.ylabel('Genre', color='white')

# Invert y-axis so the most popular genre is at the top
plt.gca().invert_yaxis()

# Dark background
plt.gca().set_facecolor('#000000')
plt.gcf().patch.set_facecolor('#000000')

# White tick labels and spines
plt.xticks(color='white')
plt.yticks(color='white')
for spine in plt.gca().spines.values():
    spine.set_color('white')

plt.grid(axis='x', color='gray', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
movies = df[df['type'] == 'Movie'].copy()
tv_shows = df[df['type'] == 'TV Show']
```

```python
movies = movies[movies['duration'].str.contains('min', na=False)]
movies['duration_num'] = movies['duration'].str.replace(' min', '', regex=False).astype(int)
plt.figure(figsize=(10,5))
plt.hist(movies['duration_num'], bins=30, color='lightblue', edgecolor='black')
plt.title('Movie Duration Distribution')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.show()
```

```python
tv_show_seasons = tv_shows['duration'].value_counts().head(10)
plt.figure(figsize=(8,4))
tv_show_seasons.plot(kind='bar', color='gold', edgecolor='black')
plt.title('Number of TV Shows by Seasons (Top 10)')
plt.xlabel('Seasons')
plt.ylabel('Number of TV Shows')
plt.xticks(rotation=45)
plt.show()
```

```python
top_directors = df[df['director'] != 'Unknown']['director'].value_counts().head(10)
top_directors.plot(kind='barh', figsize=(8,4), color='teal', edgecolor='black')
plt.title('Top 10 Directors on Netflix')
plt.xlabel('Number of Titles')
plt.gca().invert_yaxis()
plt.show()
```

```python
cast = df[df['cast'] != 'Unknown']['cast'].str.split(', ').explode()
top_actors = cast.value_counts().head(10)
top_actors.plot(kind='barh', figsize=(8,4), color='purple', edgecolor='black')
plt.title('Top 10 Actors on Netflix')
plt.xlabel('Number of Titles')
plt.gca().invert_yaxis()
plt.show()
```
